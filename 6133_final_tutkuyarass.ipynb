```python
# this is for easier usage of numpy package
import numpy as np
```

```python
# to make 100*100 random integer array
array1 = np.random.randint(0, high = 101, size=(100,100))
```

```python
array1
```

```python
# seaborn is one of package for heat map 
import seaborn as sns
```

```python
# this is default code to built up heat map
p1 = sns.heatmap(array1)
```

```python
# you can change your min and max value while constracting heatmap
p1_1= sns.heatmap(array1, vmin=0, vmax=100)
p1_1
```

```python
# you can change the color of your heat map by using cmap variable, and also linewidth and linecolor could be determined.
p1_2 = sns.heatmap(array1, cmap="YlGnBu", linewidth = 0.5, linecolor = 'white')
p1_2
```

```python
# you can determine range of your scale by using xticklabels.
p1_3 = sns.heatmap(array1, xticklabels = 5, linewidth = 1, linecolor = 'white')
p1_3
```

```python
# this is another example for xticklabels.
p1_4 = sns.heatmap(array1, xticklabels = 10, linewidth = 1, linecolor = 'white')
p1_4
```

```python
# you can remove x or y labels by defining false.
p1_5 = sns.heatmap(array1, xticklabels = 5, yticklabels = False, linewidth = 1, linecolor = 'white')
p1_5
```

```python
# this is another example for determinig max and min values of heat map.
p1_6= sns.heatmap(array1, vmin=0, vmax=50)
p1_6
```

```python
# to convert odd numbers into true, and even numbers for false we need 2 loops because we have 100*100 array.
filter_arr = np.zeros((100, 100), dtype= int)
for element in range(0, len(array1)):
    for x in range(0, len(array1[element])):
        if array1[element,x] % 2 == 0:
            filter_arr[element, x] = int(0)
        else:
            filter_arr[element,x] = int(1)
            
print(filter_arr)
array1
```

```python
# default heat map
p2 = sns.heatmap(filter_arr)
```

```python
p2_1 = sns.heatmap(filter_arr, xticklabels = False, yticklabels = False, linewidth = 1, linecolor = 'white')
p2_1
```
